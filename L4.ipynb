**Exercise 2:**

```python
#a
covid = pd.read_csv('/content/covid19_cases.csv') 
covid.head()
```

```python
#b
covid = covid.assign(
    date = lambda x: pd.to_datetime(x.dateRep)
)
covid.head()
```

```python
#c
[covid.set_index('date')]
covid.sort_index()
```

```python
#d
covid.replace(regex={'United_States_of_America':'USA','United_Kingdom':'UK'}, inplace=True)
covid[covid.countriesAndTerritories =="USA"].head()
```

```python
#e
e = covid[covid.countriesAndTerritories.isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])]
e.head()
```

```python
#f
f = e.reset_index().pivot(index='date', columns='countriesAndTerritories', values='cases').fillna(0)
f.head()
```
